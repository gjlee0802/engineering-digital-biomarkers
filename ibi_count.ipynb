```python
for id in range(1, len(dg_df)+1):
    if id == 15:
        continue
    ibi_df = pd.read_csv(f'./big-ideas-lab-glycemic-variability-and-wearable-device-data-1.1.2/{id:03}/IBI_{id:03}.csv')

    print(f'[ID: {id}] Now processing ...')

    # 데이터 처리
    ibi_df['datetime'] = pd.to_datetime(ibi_df['datetime'])
    ibi_df.set_index('datetime', inplace=True)

    # 5분 간격으로 리샘플링하고 각 구간의 개수 계산
    resampled = ibi_df[' ibi'].resample('5T').count()

    # 결과를 DataFrame으로 변환
    resampled_df = resampled.reset_index()
    resampled_df.columns = ['datetime', 'count']

    # 결과를 파일로 저장
    resampled_df.to_csv(f'./ibi_resampled_counts/ID_{id:03}_5min_counts.csv', index=False)
    
    print(f'[ID: {id}] Resampled data saved.')
```
